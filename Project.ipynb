# Remove Outlier (boxplot)

```
from matplotlib import pyplot as plt
```

```
plt.boxplot(merged_table_nm['จำนวนครั้งที่เกิดเหตุ'])
```

```
plt.boxplot(merged_table_nm['จำนวนครั้งที่เกิดเหตุ'],vert=False)
```

```
ret = plt.boxplot(merged_table_nm['จำนวนครั้งที่เกิดเหตุ'],vert=False)
```

```
ret['whiskers']
```

```
print(ret['whiskers'][0].get_xdata()) #เราปริ้นออกมาดูเพื่อหาขอบเขตมากสุดและน้อยสุดของ whiskers
print(ret['whiskers'][1].get_xdata())
```

```
no_outlier_no_missing = merged_table_nm[(merged_table_nm['จำนวนครั้งที่เกิดเหตุ']>=ret['whiskers'][0].get_xdata()[1])]
no_outlier_no_missing = no_outlier_no_missing[(no_outlier_no_missing['จำนวนครั้งที่เกิดเหตุ']<=ret['whiskers'][1].get_xdata()[1])]
```
