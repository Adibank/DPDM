```
import numpy as np
import pandas as pd
from pyod.models.copod import COPOD
import matplotlib.pyplot as plt

def nearest_pd(A):
    B = (A + A.T) / 2
    _, s, V = np.linalg.svd(B)
    H = V.T @ np.diag(s) @ V
    A2 = (B + H) / 2
    A3 = (A2 + A2.T) / 2
    if is_pd(A3):
        return A3
    spacing = np.spacing(np.linalg.norm(A))
    I = np.eye(A.shape[0])
    k = 1
    while not is_pd(A3):
        mineig = np.min(np.real(np.linalg.eigvals(A3)))
        A3 += I * (-mineig * k**2 + spacing)
        k += 1
    return A3

def is_pd(B):
    try:
        np.linalg.cholesky(B)
        return True
    except np.linalg.LinAlgError:
        return False

# จำลองข้อมูล
np.random.seed(42)
dim = 7
n = 300
alpha_test = 0.05
n_sample = int(n * (1 - alpha_test))

my_mu = np.zeros(dim)
rand_mat = np.random.uniform(0.7, 0.9, size=(dim, dim))
symm_mat = (rand_mat + rand_mat.T) / 2
np.fill_diagonal(symm_mat, 1)
cov_matrix = nearest_pd(symm_mat)
cov_matrix = np.round(cov_matrix, 2)
np.fill_diagonal(cov_matrix, 1)

data = np.random.multivariate_normal(mean=my_mu, cov=cov_matrix, size=n_sample)
df = pd.DataFrame(data, columns=[f'X.{i+1}' for i in range(dim)])

# ใช้ COPOD ตรวจจับ outlier
model = COPOD()
model.fit(df)
df['outlier'] = model.predict(df)
df['score'] = model.decision_function(df.iloc[:, :-1])

# แสดงผล 2 มิติ
plt.figure(figsize=(8, 6))
plt.scatter(df['X.1'], df['X.2'], c=df['outlier'], cmap='coolwarm', edgecolors='k')
plt.title("COPOD Outlier Detection (X.1 vs X.2)")
plt.xlabel("X.1")
plt.ylabel("X.2")
plt.grid(True)
plt.show()
```
